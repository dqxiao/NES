```python
import matplotlib.pyplot as plt
```

```python
import pickle 

def readData(folder,fname):
    f = open(folder+fname+".pickle", "rb" )
    h = pickle.load(f)
    return h 
```

```python
labels={
    0: "Training Acc",
    1: "Training Loss",
    2: "Testing Acc",
    3: "Testing Loss"
}

def title_gen(folder):
    fs=folder.split("/")
    return "{}-{}".format(fs[1],fs[2])

def compare(index):
    
    for m in methods:
        h=readData(folder,m)
        plt.plot(h[start:end,index],label=m)
#         else:
#             plt.plot(h[start:end,index],label=m)
    if "Loss" in  labels[index]:
        plt.yscale('log')
    plt.xlabel('Effective Pass')
    plt.ylabel(labels[index])
    plt.legend(loc="best")
    plt.title(title_gen(folder))
    plt.show()
```

```python
folder="./CIFAR10/CNN/"
methods=["SGDM-0.99-NN","SGDM-0.9-NN","SGDM_OES-0.7-0.99-NN","SGDM_PEPG-0.8-0.99-NN"]
```
