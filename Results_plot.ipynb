```python
import matplotlib.pyplot as plt
```

```python
import pickle 

def readData(folder,fname):
    f = open(folder+fname+".pickle", "rb" )
    h = pickle.load(f)
    return h 
```

```python

start,end =0, 100
```

```python
labels={
    0: "Training Loss",
    1: "Training Acc",
    2: "Testing Loss",
    3: "Testing Acc"
}

def title_gen(folder):
    fs=folder.split("/")
    return "{}-{}".format(fs[1],fs[2])

def compare(index):
    
    for m in methods:
        h=readData(folder,m)
        plt.plot(h[start:end],label=m)
#         else:
#             plt.plot(h[start:end,index],label=m)
    plt.yscale('log')
    plt.xlabel('Effective Pass')
    plt.ylabel(labels[index])
    plt.legend(loc="best")
    plt.title(title_gen(folder))
    plt.show()
```

```python
folder="./MNIST/MLP/"
methods=["PEPG-1-0.0","PEPGVar-1-0.1","PEPGVar-1-0.2","PEPGVar-1--0.1","PEPGVar-1-0.01","PEPGVar-1-0.15"]
# methods=["slr0.1PEPGVar-1-0.01","slr0.1PEPG-1-0.0","slr0.1PEPGVar-1-0.02"]
# methods=["slr0.1BP-PEPG-1-0.0","slr0.1BP-PEPGVar-1-0.01"]
start, end = 0, 200 
h=readData(folder, methods[0])
eh=readData(folder,methods[1])
index=-1
print("{}->{}".format(h[index],eh[index]))
compare(0)
```
